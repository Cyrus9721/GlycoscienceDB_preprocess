```python
import numpy as np
import pandas as pd
import os
pd.set_option('display.max_rows', 2000)
pd.set_option('display.max_columns', 2000)
```

```python
data_dir = 'experimental_data_combined/FullyAnnotatedPDB_pdb_labeled_combined_add_temperature_reformulate_pdb/'
new_data_dir = 'experimental_data_combined/FullyAnnotatedPDB_pdb_labeled_combined_add_temperature_reformulate_pdb_outlier_removal/'

out_dir_1 = 'experimental_data_combined/FullyAnnotatedPDB_pdb_labeled_combined_add_temperature_reformulate_pdb_outlier_removal_second_round/'
```

```python
# given specific glycan's outlier carbon number, original values, new value, reassign its value
def switch_values(f_name, temp_f, carbon_value, carbon_number, new_carbon_value):
    
    out_df = temp_f.copy()
    
    if isinstance(carbon_value, list):
        for i in range(len(carbon_value)):
            
            current_val = carbon_value[i]
            
            current_num = carbon_number[i]
            
            new_val = new_carbon_value[i]
            
            if out_df.loc[(out_df['Atom_Type'] == current_num) & (out_df['labels'] == current_val)].empty:
                print(f_name, current_num, 'with value of:', current_val, 'does not exist')
            elif len(out_df.loc[(out_df['Atom_Type'] == current_num) & (out_df['labels'] == current_val)]['labels']) > 1:
                print(f_name, current_num, 'with value of:', current_val, 'has more than value need inspect')
            else:
                out_df.loc[(out_df['Atom_Type'] == current_num) & (out_df['labels'] == current_val), ['labels']] = new_val
            
    else:
        current_val = carbon_value
            
        current_num = carbon_number

        new_val = new_carbon_value

        if out_df.loc[(out_df['Atom_Type'] == current_num) & (out_df['labels'] == current_val)].empty:
            
            print(f_name, current_num, 'with value of:', current_val, 'does not exist')
            
        elif len(out_df.loc[(out_df['Atom_Type'] == current_num) & (out_df['labels'] == current_val)]['labels']) > 1:
            
            print(f_name, current_num, 'with value of:', current_val, 'has more than value need inspect')
            
        else:
            out_df.loc[(out_df['Atom_Type'] == current_num) & (out_df['labels'] == current_val), ['labels']]= new_val

    
    return out_df
```

```python
# drop all the values of pdb by monosaccharide name 
def drop_value_by_name(temp_df, mono_name):
    
    temp_pdb_df = temp_df.copy()
    
    temp_pdb_df.loc[temp_pdb_df['origin_mono'] == mono_name, ['labels']] = -1.00
    
    return temp_pdb_df
```

```python
# drop all the values of a 
def drop_value_by_mono(f, temp_pdb, mono_number):
    
    out_df = temp_pdb.copy()
    
    out_df.loc[out_df['residual'] == mono_number, ['labels']] = -1.00
    
    return out_df
```

```python
orig_files = os.listdir(data_dir)
for f in orig_files:
    temp_pdb = pd.read_csv(os.path.join(data_dir, f))
        
    
    # training outlier
    if f == 'a-D-Glcp-(1-3)-a-L-Galp.pdb.csv':
        modified_pdb = switch_values(f, temp_pdb, 177.11, 'C6', -1)
    
    elif f == 'b-D-Manp.pdb.csv':
        modified_pdb = switch_values(f, temp_pdb, 176.82, 'C6', -1)
        
    elif f == 'Repeat-4)-a-L-IdopA-(1-4)-a-D-GlcpNAc-(1-.pdb.csv':
        modified_pdb = switch_values(f, temp_pdb, 175.1, 'C6', -1)
        
        
        
    elif f == 'b-D-Glcp-(1-4)-b-D-GlcpA-(1-3)-a-D-GalpA2SO3-(1-4)-b-D-Galp.pdb.csv':
        modified_pdb = switch_values(f, temp_pdb, 178.6, 'C6', 65)
    
    elif f == 'b-D-Glcp-(1-4)-b-D-GlcpA-(1-3)-a-D-GalpA-(1-4)-b-D-Galp.pdb.csv':
        modified_pdb = switch_values(f, temp_pdb, 178.6, 'C6', 65)
    
    elif f == 'DB17331.pdb.csv':
        modified_pdb = switch_values(f, temp_pdb, [101.750, 56.84, 36.37, 131.39, 131.54, 117.82], 
                                     ['C1', 'C2', 'C3', 'C4', 'C5','C6'], [-1, -1, -1, -1, -1, -1])
    
    elif f == 'b-D-GalpANAc.pdb.csv':
        modified_pdb = switch_values(f, temp_pdb, 61.89, 'C6', 175)
        
    elif f == 'DB26961.pdb.csv':
        modified_pdb = switch_values(f, temp_pdb, 69, 'C1', 96.2)
        
    elif f == 'DB26879.pdb.csv':
        modified_pdb = drop_value_by_mono(f, temp_pdb, 7)
        
#     elif f == 'DB26910.pdb.csv':
#         modified_pdb = switch_values(f, temp_pdb, 70.3, 'C2', -1)
        
#         prev_pdb = modified_pdb
        
    elif f == 'a-D-Glcp-(1-3)-b-L-Galp.pdb.csv':
        modified_pdb = switch_values(f, temp_pdb, 177.11, 'C6', -1)
    
    elif f == 'a-D-GlcpNAc-(1-4)-a-D-GlcpNAc-(1-4)-a-D-Kdop.pdb.csv':
        modified_pdb = switch_values(f, temp_pdb, 174.1, 'C6', -1)
        
    elif f == 'DB26404.pdb.csv':
        modified_pdb = switch_values(f, temp_pdb, 61.9, 'C1', -1)
    
#     elif f == 'DB26431.pdb.csv':
#         pass
    
#     elif f == 'DB26959.pdb.csv':
#         prev_pdb = temp_pdb
    
    
    # Third round outlier 
    
    elif f == 'b-D-Glcp-(1-4)-b-D-GlcpA-(1-3)-a-D-GalpA-(1-4)-a-D-Galp.pdb.csv':
        modified_pdb = switch_values(f, temp_pdb, 178.6, 'C6', 65)
        
    
    elif f == 'b-D-Glcp-(1-4)-b-D-GlcpA-(1-3)-a-D-GalpA2SO3-(1-4)-a-D-Galp.pdb.csv':
        modified_pdb = switch_values(f, temp_pdb, 178.6, 'C6', 65)
        
        
    elif f == 'DB26479.pdb.csv':
        modified_pdb = temp_pdb.copy()
        modified_pdb.loc[(modified_pdb['Atom_Type'] == 'C6') & (modified_pdb['residual'] == 5), ['labels']] = 65
        modified_pdb = switch_values(f, modified_pdb, 5.16, 'H3', -1)
        modified_pdb = switch_values(f, modified_pdb, 4.84, 'H2', -1)
        
        
    elif f == 'DB26378.pdb.csv':
        modified_pdb = switch_values(f, temp_pdb, 170.0, 'C6', -1)
        modified_pdb = switch_values(f, modified_pdb, 146.5, 'C5', -1)
        modified_pdb = switch_values(f, modified_pdb, 107.5, 'C4', -1)
        modified_pdb = switch_values(f, modified_pdb, 5.723, 'H4', -1)
        modified_pdb = switch_values(f, modified_pdb, 2.931, 'H2', -1)
        modified_pdb = switch_values(f, modified_pdb, 2.863, 'H2', -1)
        modified_pdb = switch_values(f, modified_pdb, 3.052, 'H2', -1)
        
        
        
    elif f == 'a-L-Rhap-(1-3)-b-D-Glcp.pdb.csv':
        modified_pdb = switch_values(f, temp_pdb, 18, 'C4', -1)
        modified_pdb = switch_values(f, modified_pdb, [61.6, 85.0, 68.4, -1. , -1. , -1. ], 
                                     ['C1', 'C2', 'C3', 'C4', 'C5','C6'], [-1, -1, 85.0, 68.4, -1, 61.6])
        
        modified_pdb = switch_values(f, modified_pdb, 1.16, 'H4', -1)
        
    
    # Fourth round hydrogen
    elif f == 'DB22529.pdb.csv':
        
        modified_pdb = drop_value_by_name(temp_pdb, 'galfoac')
        
        
    elif f == 'DB26963.pdb.csv':
        modified_pdb = drop_value_by_name(temp_pdb, 'galfoac')
        
    
    elif f == 'b-L-Fucp-(1-2)-b-L-Rhap.pdb.csv':
        modified_pdb = switch_values(f, temp_pdb, 1.28, 'H5', -1)
        
    elif f == 'DB26467.pdb.csv':
        prev_pdb = temp_pdb.copy()
        
        modified_pdb = switch_values(f, temp_pdb, 149.2, 'C5', -1)
        modified_pdb = switch_values(f, modified_pdb, 108.3, 'C4', -1)
        
        current_pdb = modified_pdb.copy()
    elif f == 'DB26383.pdb.csv':
        modified_pdb = switch_values(f, temp_pdb, 50.9, 'C3', -1)
        
    elif f == 'DB26708.pdb.csv':
#         modified_pdb = switch_values(f, temp_pdb, 50.9, 'C3', -1)
        modified_pdb.loc[(modified_pdb['Atom_Type'] == 'C2') & (modified_pdb['residual'] == 1), ['labels']] = -1
        
    
    elif f == 'a-L-Rhap-(1-4)-a-D-Glcp-(1-3)-b-L-Rhap-(1-4)-b-D-Glcp-(1-4)-a-D-Glcp-(3-1)-a-L-Rhap.pdb.csv':
        
        
        switch_values_list = ['C1', 'C2', 'C3', 'C4', 'C5', 'C6', 'H1', 'H2', 'H3', 'H4', 'H5']

        for i in switch_values_list:

            res1_val = temp_pdb.loc[(temp_pdb['residual'] == 1) & (temp_pdb['Atom_Type'] == i), :]['labels'].values[0]

            res2_val = temp_pdb.loc[(temp_pdb['residual'] == 2) & (temp_pdb['Atom_Type'] == i), :]['labels'].values[0]

            temp_pdb.loc[(temp_pdb['residual'] == 1) & (temp_pdb['Atom_Type'] == i), ['labels']] = res2_val
            temp_pdb.loc[(temp_pdb['residual'] == 2) & (temp_pdb['Atom_Type'] == i), ['labels']] = res1_val
        
        modified_pdb = temp_pdb
        
        
        
    elif f == 'b-D-Quip-(1-6)-b-D-Galp-(1-6)-b-D-GlcpNAc-(1-6)-b-D-Galp-(1-3)-b-D-GlcpNAc.pdb.csv':
        modified_pdb = switch_values(f, temp_pdb, 58.1, 'C4', -1)
#         prev_pdb = modified_pdb
    
    # end Fourth round 
    else:
        modified_pdb = temp_pdb
        
    if 'KDO' in temp_pdb['Monosaccharide_belong'].values:
#         continue
        modified_pdb.to_csv(os.path.join('experimental_data_combined/kdo', f), index = False)
        
    elif 'SIA' in temp_pdb['Monosaccharide_belong'].values:
        
        continue
    
    elif f == 'b-D-Glcp-(1-6)-b-D-Glcp-(1-6)-b-D-GlcpNAc-(1-2)-a-D-Glcp.pdb.csv':
        
        continue
        
    else:
        
#         modified_pdb = temp_pdb
        
        modified_pdb.to_csv(os.path.join(out_dir_1, f), index = False)
```
