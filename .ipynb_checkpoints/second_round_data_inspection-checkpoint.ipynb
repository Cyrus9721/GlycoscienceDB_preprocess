```python
atom_type_list = []
for pdb_name in labeled_pdb_files:
    pdb_f = pd.read_csv(os.path.join(labeled_pdb_dir, pdb_name))
    
    current_atom_list = pdb_f['Atom_Type'].values
    current_shift_values = pdb_f['labels'].values
    
    atom_type_list.extend(list(current_atom_list))
    
    for i in range(len(current_atom_list)):
        if current_atom_list[i] == 'C1':
            C1_list.append(current_shift_values[i])
            C1_list_pdb_name.append(pdb_name)
            
        if current_atom_list[i] == 'C2':
            C2_list.append(current_shift_values[i])
            C2_list_pdb_name.append(pdb_name)
            
        if current_atom_list[i] == 'C3':
            C3_list.append(current_shift_values[i])
            C3_list_pdb_name.append(pdb_name)
            
        if current_atom_list[i] == 'C4':
            C4_list.append(current_shift_values[i])
            C4_list_pdb_name.append(pdb_name)
            
        if current_atom_list[i] == 'C5':
            C5_list.append(current_shift_values[i])
            C5_list_pdb_name.append(pdb_name)
            
        if current_atom_list[i] == 'C6':
            C6_list.append(current_shift_values[i])
            C6_list_pdb_name.append(pdb_name)
            

        if current_atom_list[i] == 'H1':
            H1_list.append(current_shift_values[i])
            H1_list_pdb_name.append(pdb_name)
            
        if current_atom_list[i] == 'H2':
            H2_list.append(current_shift_values[i])
            H2_list_pdb_name.append(pdb_name)
            
        if current_atom_list[i] == 'H3':
            H3_list.append(current_shift_values[i])
            H3_list_pdb_name.append(pdb_name)
            
        if current_atom_list[i] == 'H4':
            H4_list.append(current_shift_values[i])
            H4_list_pdb_name.append(pdb_name)
            
        if current_atom_list[i] == 'H5':
            H5_list.append(current_shift_values[i])
            H5_list_pdb_name.append(pdb_name)
            
        if current_atom_list[i] == 'H6':
            H6_list.append(current_shift_values[i])
            H6_list_pdb_name.append(pdb_name)
            
        if current_atom_list[i] == 'H61':
            H61_list.append(current_shift_values[i])
            H61_list_pdb_name.append(pdb_name)
            
        if current_atom_list[i] == 'H62':
            H62_list.append(current_shift_values[i])
            H62_list_pdb_name.append(pdb_name)

atom_type_list = np.array(sorted(list(set(atom_type_list))))
```

##### 2, first corrections

```python
for pdb_name in labeled_pdb_files:
    pdb_f = pd.read_csv(os.path.join(labeled_pdb_dir, pdb_name))
    
    current_atom_list = pdb_f['Atom_Type'].values
    current_shift_values_list = pdb_f['labels'].values
    
    current_labeled_list = np.zeros(len(current_atom_list)).astype(int)
    
    for i in range(len(current_atom_list)):
        
        current_atom = current_atom_list[i]
        current_shift_value = current_shift_values_list[i]
        
        if current_atom in carbon_list:
            
            if current_shift_value < 10:
                pass
            else:
                current_labeled_list[i] = 1
#                 print(current_shift_value)
        
        if current_atom in hydrogen_list_1_6:
            
            if current_shift_value == -1:
                pass
            else:
                current_labeled_list[i] = 2
#                 print(current_shift_value)
```
