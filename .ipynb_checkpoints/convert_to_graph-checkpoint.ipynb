```python
def experiment_process_single_pdb(f):
    temp_shift = []
    for f_l in f:
        if 'HETATM' in f_l:
            temp_shift.append(f_l)

    new_shift = []
    for i in range(len(temp_shift)):
        temp_l = temp_shift[i].split(' ')
        temp_l_s = [i for i in temp_l if i != '']
        new_shift.append(temp_l_s[0:8])
    return new_shift
```

```python
def handle_missing_residual_pdb(new_f):
    new_f_handle_missing_residual = []
    for i in range(len(new_f)):
        temp_l = new_f[i]
        try:
            fourth_element = int(temp_l[3])
            temp_l.insert(3, 'Missing monosaccharide')
            temp_l.pop()
            new_f_handle_missing_residual.append(temp_l)
        except:
            new_f_handle_missing_residual.append(temp_l)
    return new_f_handle_missing_residual
```

##### 3, Asigning residuals to csv atoms 

```python
def assign_residuals(temp_csv, H_initial = ['H11','H12','H1'], C_initial = ['C1']):
    res_num = 0
    c_res = 0
    h_res = 0
    temp_residual_list = []
    prev_atom = ''
    prev_connection = 'some text'
    for i in range(len(temp_csv)):
        
        current_atom = temp_csv.loc[i, ].Atom
        
        current_connection = temp_csv.loc[i, ].Connection
        
        
        if current_atom in H_initial:
            
            if (prev_atom in H_initial):
                pass
            
            else:
                h_res += 1
                res_num = h_res
                
        elif current_atom in C_initial:
            
            if (prev_atom in C_initial):
                pass
            
            else:
                c_res += 1
                res_num = c_res
                
        temp_residual_list.append(res_num)
        
        prev_atom = current_atom
        
        prev_connection = current_connection
        
    temp_csv['residual_num'] = temp_residual_list
    
    return temp_csv
```

##### 5, Correct residuals, some of the residuals are labled in a wrong way

```python
# case 1: none glycan wrong labeled
def correct_csv_attached_none_glycan(temp_df, monasaccharide_to_be_deleted = ''):
    # for some glycans 
    # 1 -> x, 
    # 2 -> 1
    # 3 -> 2 ....
    temp_residual_num = temp_df['residual_num'].values
    
    temp_residual_num[temp_residual_num == 0] = 1
    
    temp_residual_num_correct = temp_residual_num - 1
    
    temp_residual_num_correct[temp_residual_num_correct == 0] = np.max(temp_residual_num_correct) + 1
    
    temp_df['residual_num'] = temp_residual_num_correct
    
    df = temp_df
    # delete attached non-monosaccharide residuals
    if monasaccharide_to_be_deleted:
        df = df[df.Residual != monasaccharide_to_be_deleted]
        df.index = range(len(df))
    return df
```

```python
def delete_none_glycan(temp_df, monasaccharide_to_be_deleted = ''):
    df = temp_df.copy()
    # delete attached non-monosaccharide residuals
    if monasaccharide_to_be_deleted:
        df = df[df.Residual != monasaccharide_to_be_deleted]
        df.index = range(len(df))
    return df
```

```python
def correct_csv_attached_zero_labeled(temp_df, monasaccharide_to_be_deleted = ''):
    # for some glycans starts with H12, correct their labels
    temp_residual_num = temp_df['residual_num'].values
    
    temp_residual_num[temp_residual_num == 0] = 1
    
    temp_df['residual_num'] = temp_residual_num
    
    return temp_df
```

```python
# case 2: Fruf
def correct_csv_attached_Fruf(temp_df):
    
    temp_residual_mono = temp_df['Residual'].values
    
    mono_index_list = []
    
    # Delete FRUF terms since they are not complete monosaccharide and poor labeled 
    
    for i in range(len(temp_residual_mono)):

        if 'FRUF' in temp_residual_mono[i]:
            pass
        else:
            mono_index_list.append(i)
    
    df = temp_df.loc[mono_index_list, :]
    df.index = range(len(df))
    
    # modify residuals
    
    temp_residual_num = df['residual_num'].values
    
    min_res = np.min(temp_residual_num)
    
    temp_residual_num -= (min_res - 1)
    
    df['residual_num'] = temp_residual_num
    return df
```

```python
# case 4, some glycans do not contain 'H1 or H11' in the initials

def correct_assigned_by_connection(temp_df):
    H_initial = ['H11','H1']
    C_initial = ['C1']
    prev_connection = 'some text'
    prev_num = 99
    
    df = temp_df.copy()
    
    changed_index_list= []
    
    changed_values_list = []
    
    for i in range(len(df)):
        current_connection = df.loc[i, ].Connection
        current_num = df.loc[i, ].residual_num
        current_index = '-99' 
        if (prev_connection != current_connection) and (prev_connection != 'some text') and (prev_num == current_num):
#             print(i)
            current_index = i 
            current_residual = df.loc[i, ].Residual
            
            if (changed_index_list) and (changed_index_list[-1] + 1) == current_index:
#                 print(current_index)
                current_changed_value = changed_values_list[-1] + 1
            else:
                current_changed_value = df.loc[current_index, ].residual_num + 1
            
            
            changed_index_list.append(current_index)
            changed_values_list.append(current_changed_value)
            
            
            while ((current_index < len(df)-1) and (df.loc[current_index + 1, ].Connection == df.loc[current_index, ].Connection) and 
                  (df.loc[current_index + 1, ].Residual == df.loc[current_index, ].Residual) and 
                  (df.loc[current_index + 1, ].Atom not in H_initial) and 
                  (df.loc[current_index + 1, ].Atom not in C_initial)):
                
                changed_index_list.append(current_index)
                changed_values_list.append(current_changed_value)
                
            
                current_index += 1
                
                
#                 print(current_index)
            
            changed_index_list.append(current_index)
            changed_values_list.append(current_changed_value)
            

        prev_connection = current_connection
        prev_num = current_num
    
    df.loc[changed_index_list, ['residual_num']] = np.array(changed_values_list).astype(int)
    
    
    return df
```

```python
pdb_dir = 'experimental_data/FullyAnnotatedPDB_pdb/'
correct_csv_dir = 'experimental_data/FullyAnnotatedPDB_label_corrected/'
out_dir_labeled_pdb = 'experimental_data/FullyAnnotatedPDB_pdb_labeled/'
pdb_files = os.listdir(pdb_dir)
correct_labels = os.listdir(correct_csv_dir)

carbon_list = ['C1', 'C2', 'C3', 'C4', 'C5', 'C6']
hydrogen_list_1_6 = ['H1', 'H2', 'H3', 'H4', 'H5', 'H6']
for i in range(len(pdb_files)):
    temp_pdb = pd.read_csv(os.path.join(pdb_dir, pdb_files[i]))
    temp_label_csvs = pd.read_csv(os.path.join(correct_csv_dir, correct_labels[i]))
    
    temp_labels = np.repeat(-1.0, len(temp_pdb))
    
    temp_all_atoms = temp_pdb['Atom_name'].values
    temp_all_residual = temp_pdb['Residual_num'].values
    
    
    for j in range(len(temp_pdb)):
        current_atom = temp_all_atoms[j]
        current_residual = temp_all_residual[j]
        
        if current_atom in carbon_list or current_atom in hydrogen_list_1_6:
        
            try:
                temp_labels[j] = temp_label_csvs.loc[(temp_label_csvs['Atom'] == current_atom) & 
                                                     (temp_label_csvs['residual_num'] == current_residual)].Shift.values[0]
            except:
                pass
    temp_pdb['labels'] = temp_labels
#     print(temp_labels)
    temp_pdb.columns = ['Atom','Atom_Num','Atom_Type','Monosaccharide_belong','residual','x','y','z','labels']
    temp_pdb.to_csv(os.path.join(out_dir_labeled_pdb, 'labeled' + pdb_files[i]))
```
