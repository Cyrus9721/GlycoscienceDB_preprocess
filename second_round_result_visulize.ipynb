```python

def replace_kdo(arr):
    new_arr = []
    for i in arr:
        if 'KDO' in i:
            new_arr.append(i.replace('_KDO', ''))
        else:
            new_arr.append(i)
    return np.array(new_arr)
```

```python
def plot_shift(df, ax, i, j, set_title_name = '', sep = 5, xlab_name = '', ylab_name = '', 
               color = 'grey', name = None, color_plate_dict = None, ms = 2, font_size = 12):
    acctual_shit_1 = df.loc[df['atom_type'] == name]['act'].values
    predict_shit_1 = df.loc[df['atom_type'] == name]['pred'].values
    low_limit = np.min([np.min(np.abs(acctual_shit_1)), np.min(np.abs(predict_shit_1))]) * 0.6
    up_limit = np.max([np.max(np.abs(acctual_shit_1)), np.max(np.abs(predict_shit_1))]) * 1.2
    
    mse = sklearn.metrics.mean_squared_error(acctual_shit_1, predict_shit_1)
    rmse = math.sqrt(mse)  
    rmse = np.round(rmse, 3)
    
    set_title_name = 'RMSE = '+str(rmse)
    
    sns.set_theme()
    sns.set_style("whitegrid")
    x = np.linspace(low_limit, up_limit)
    ax[i, j].plot(x, x, color=color, alpha=0.25, zorder=0)
    
#     ax[i, j].scatter(acctual_shit_1, predict_shit_1, s = 20, label = 'A', linewidths=0, color=color)
    ax[i, j].plot(predict_shit_1, acctual_shit_1, marker='o', linestyle='', ms=ms, label=name, color = color_plate_dict[name])
    
    ax[i, j].ticklabel_format(style="sci", scilimits=(-4, 4))
    ax[i, j].axis('square')
    
    ax[i, j].set_title(f'{set_title_name}', size = font_size)
    ax[i, j].set_ylim(low_limit, up_limit)
    ax[i, j].set_xlim(low_limit, up_limit)
#     ax[i, j].grid()
    grid_x_ticks = np.arange(low_limit, up_limit, (up_limit - low_limit) / sep)
    grid_y_ticks = np.arange(low_limit, up_limit, (up_limit - low_limit) / sep)
    
    ax[i, j].set_xlabel(xlab_name, loc="center", size = font_size)
    ax[i, j].set_ylabel(ylab_name, loc="center", size = font_size)
#     ax[i, j].yaxis.set_label_coords(x = y_lab_xcoordinate, y = 0.5)
    
    ax[i, j].set_xticks(grid_x_ticks , minor=True)
    
    
    ax[i, j].set_yticks(grid_y_ticks , minor=True)

    ax[i, j].grid(which='both')

    ax[i, j].grid(which='minor', alpha=1, linestyle='-')
    ax[i, j].legend(fontsize = 12, markerscale=3)
```

```python
sns.set_theme(style="whitegrid", font_scale=1)

lab_title_font_size = 16
figure, axis = plt.subplots(2, 3, figsize=(13, 8), constrained_layout=True)
plot_shift(df_hydrogen_all, axis, 0, 0, set_title_name = '', sep = 5, xlab_name = 'Pred. Shift', 
           ylab_name = 'Act. Shift', color = '#679399', name='H1', 
           color_plate_dict=colors_hydrogen, ms = 5, font_size=lab_title_font_size)

plot_shift(df_hydrogen_all, axis, 0, 1, set_title_name = '', sep = 5, xlab_name = 'Pred. Shift', 
           ylab_name = '', color = '#679399', name='H2', 
           color_plate_dict=colors_hydrogen, ms = 5, font_size=lab_title_font_size)

plot_shift(df_hydrogen_all, axis, 0, 2, set_title_name = '', sep = 5, xlab_name = 'Pred. Shift', 
           ylab_name = '', color = '#679399', name='H3', 
           color_plate_dict=colors_hydrogen, ms = 5, font_size=lab_title_font_size)

plot_shift(df_hydrogen_all, axis, 1, 0, set_title_name = '', sep = 5, xlab_name = 'Pred. Shift', 
           ylab_name = '', color = '#679399', name='H4', 
           color_plate_dict=colors_hydrogen, ms = 5, font_size=lab_title_font_size)

plot_shift(df_hydrogen_all, axis, 1, 1, set_title_name = '', sep = 5, xlab_name = 'Pred. Shift', 
           ylab_name = '', color = '#679399', name='H5', 
           color_plate_dict=colors_hydrogen, ms = 5, font_size=lab_title_font_size)

plot_shift(df_hydrogen_all, axis, 1, 2, set_title_name = '', sep = 5, xlab_name = 'Pred. Shift', 
           ylab_name = '', color = '#679399', name='H6', 
           color_plate_dict=colors_hydrogen, ms = 5, font_size=lab_title_font_size)
```
